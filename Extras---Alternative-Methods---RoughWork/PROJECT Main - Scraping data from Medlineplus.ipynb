```python
A= soup.find('ul', class_='withident breaklist')
```

```python
A
```

```python
links = A.find_all('a')
```

```python
links
```

```python
 for i in links:
    print(i.text.strip(),i['href'])
```

```python
Gene_name = [i.text.strip() for i in links]
print(Gene_name)  # list of gene_name
```

```python
x=[]      # list of links/urls for each gene
for i in links:
     y=  (i['href'])
     if y:
         x.append(y)
```

```python
print(x)
```

```python
for i in x[:5]:
    page2 = requests.get(i)
    soup = BeautifulSoup(page2.text,'html')
    cond = soup.find_all('div', class_='mp-content mp-exp exp-full')
    conditions = []
    for j in cond:
        z = j.find("h3")
        if z:
            z_text = z.text.strip()
            if z_text:
                conditions.append(z_text)
    print(conditions)
```

```python
gene_condition = []
for g,i in zip(Gene_name,x[:5]):
    page2 = requests.get(i)
    soup = BeautifulSoup(page2.text,'html')
    cond = soup.find_all('div', class_='mp-content mp-exp exp-full')
    conditions = []
    for j in cond:
        z = j.find("h3")
        if z:
            z_text = z.text.strip()
            if z_text:
                conditions.append(z_text)
    gene_condition.append({
        "gene": g,
        "conditions": conditions
    })
```

```python
print(gene_condition)
```

```python
with open("genes+its_condition.json", "w", encoding="utf-8") as f:
    json.dump(gene_condition, f, ensure_ascii=False, indent=2)

print("Done Brother!!!!!!!!!")
```
